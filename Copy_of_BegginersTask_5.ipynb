### Check if the data contain duplicated values or not **and** if it has ,remove these values

```python
automobile.duplicated().sum()
```

```python
auto_dropdup = automobile.drop_duplicates()
auto_dropdup
```

```python
auto_dropdup.duplicated().sum()
```

### Transform (normalized-losses) from object to numeric

```python
auto_dropdup['normalized-losses'] = pd.to_numeric(auto_dropdup['normalized-losses'], errors='coerce')

# auto_dropdup['normalized-losses'].astype(float)
auto_dropdup["normalized-losses"].describe()
```

### Replace the missing values in each columns

```python
auto_dropdup = auto_dropdup.fillna(auto_dropdup.mean())
auto_dropdup
```

### Save all changes in new excel file called <font color='blue'>cleaned data

```python
auto_dropdup.to_csv("cleaned_data.csv")
```
