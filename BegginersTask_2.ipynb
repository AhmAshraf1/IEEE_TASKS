# <font color='red'> 4) Write a function that takes any variable "n" and return its factorial: </font>

##### **hint:** factorial of 5 = 5 x 4 x 3 x 2 x **1**

```python
def factorial(n):
    fact = 1
    for i in range(n):
        fact *= (n - i)
    return fact


n = int(input())
if n >= 0:
    print("factorial of " + str(n) + " =", factorial(n))
else:
    print("n must be positive number")
```

# <font color='red'> 6) Write a Python function that accepts a string and counts the number of upper and lower case letters. </font>
* ## Sample String : "Do not Postpone today's Work To tomoRRow"

# Expected Output :
* `No. of Upper case characters :  6`
* `No. of Lower case Characters :  27 `







```python
st = input()
lower_counter = 0
upper_counter = 0
for letter in st:
    if ord(letter) in range(65,90):
        upper_counter += 1
    elif ord(letter) in range(97,122):
        lower_counter += 1
print("No. of Upper case characters : ", upper_counter)
print("No. of Lower case characters : ", lower_counter)
```

# <font color='red'> 8) Your input is corresponding to a person's age (in days). print his age in years, months and days, followed by its respective message "years", "months" and "days" . </font>

# * `Input :  400`
# * `Output:  1 year  1 months  5 days`

```python
age = int(input("Age in Days: "))
if age >= 365:
    years = int(age / 365)
    months = int((age - (years * 365)) / 30)
    days = int(((age - (years * 365)) - (months*30)))
elif age >= 30:
    months = int(age / 30)
else:
    days = int(age)
print(years, "years", months, "months", days, "days")
```
